```python
operations_and_dataset = pd.merge(operations, data_set);
all_data = pd.merge(pd.merge(operations_and_dataset, types), codes);
```

```python
from datetime import datetime, timedelta
def f(x):
    start_date = datetime.today()
    days, time = x.split(' ')
    h, m, s = time.split(':')
    new_date = start_date - timedelta(days=int(days)) - timedelta(seconds=int(h)*60*60+int(m)*60+int(s))
    return new_date

all_data['datetime'] = all_data['datetime'].apply(lambda x: f(x))
all_data['datetime']
```
